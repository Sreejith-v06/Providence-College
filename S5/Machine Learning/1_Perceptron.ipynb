```python
#Training..........
#1.Load the Data
import pandas as pd
file=pd.read_excel("/irisexcel.xlsx")
x=file[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
y=file["species"]
#Cross Validation
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest=train_test_split(x, y, train_size=0.8)
#2.Load the MODEL
from sklearn.linear_model import Perceptron
model=Perceptron()
#3.Fit the data into model
model.fit(xtrain,ytrain)
#validation Testing
op2=model.predict(xtest)
from sklearn.metrics import accuracy_score
score=accuracy_score(ytest, op2)
print("Accuracy=", score*100)
#Testing...........
op1=model.predict([[1.5, 2.5, 3.5, 4.5]])
print(op1)
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm=confusion_matrix(ytest, op2)
cmd=ConfusionMatrixDisplay(confusion_matrix=cm,
                           display_labels=model.classes_)
cmd.plot()

```
